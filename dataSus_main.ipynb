```python
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
import pandas as pd #1ª célula

dados = pd.read_csv('raw_data/sih_cnv_qiuf125519179_34_5_225.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',')
# pd.options.display.float_format = '{:.2f}'.format 
  
dados
```

```python
dados.plot(x='Unidade da Federação', y='2008/Ago')
```

```python
axis = dados.plot(x='Unidade da Federação', y='2024/Dez', kind='bar')
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
axis.set_xticklabels(axis.get_xticklabels(), rotation=45, ha='right') # Legendas na Diagonal

plt.title("Valor por Unidade da Federação")
```
